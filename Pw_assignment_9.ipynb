#1. What is a parameter?

Ans:
A parameter is a value learned by a machine learning model during training.
Example: In linear regression

             y=mx+c

   m and c are parameters learned from data

#5. How does loss value help in determining a good model?

Ans:
Loss value shows how wrong the model’s predictions are.

Low loss = good model

High loss = poor model

Example: MSE = 0.5 is better than MSE = 200

#15. What is causation? Difference between correlation vs causation?

Ans:

Causation

One variable directly affects the other.

Difference

Correlation = relationship

Causation = cause-effect relation

Example:
Ice cream sales ↑ and drowning cases ↑
→ correlated
But ice cream does not cause drowning

#16. What is an Optimizer? Types?

Ans:

Optimizer updates model parameters to reduce loss.

Types:
✔ SGD

Updates weights using gradient descent.

✔ Adam

Combination of momentum + RMSProp (faster and widely used).

✔ RMSProp

Uses moving average of gradients.
